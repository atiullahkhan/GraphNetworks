```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
citationsNetwork = nx.read_edgelist('Citation_Links_cmp.txt', create_using = nx.DiGraph(), nodetype = int)
print(nx.info(citationsNetwork))
```

```python
citationsNetwork.nodes()
```

```python
degreeCentrality = nx.degree_centrality(citationsNetwork)
degreeCentrality
```

```python
# Using readlines()
citationDates = open('Citation_Dates_cmp.txt', mode = "r", encoding = "utf-8")
Lines = citationDates.readlines()
 
citationDatesDict = {}
# Strips the newline character
for line in Lines:
    citationDatesDict[line.split()[0]] = line.split()[1]
citationDatesDict
```

```python
citationDatesDict[str(9912240)]
```

```python
for paper, centralityValue in degreeCentrality.items():
    citationsNetwork.nodes[paper]['Degree'] = centralityValue
    try:
        citationsNetwork.nodes[paper]['Citation_Date'] = citationDatesDict[str(paper)]
    except:
        pass
```

```python
nx.density(citationsNetwork)
```

```python
citationsNetwork.degree()
```

```python
count = 0
for paper, degree in citationsNetwork.degree():
    if degree == 0:
        count+=1
count
```

```python
degreeCentrality.values()
```

```python
maxCentrality = max(degreeCentrality.values())
```

```python
for paper in citationsNetwork.nodes():
    if citationsNetwork.nodes[paper]['Degree'] >= maxCentrality * 0.9:
        citationsNetwork.nodes[paper]['Importance'] = 600
        citationsNetwork.nodes[paper]['Colour'] = 'green'
    elif citationsNetwork.nodes[paper]['Degree'] >= maxCentrality * 0.2:
        citationsNetwork.nodes[paper]['Importance'] = 350
        citationsNetwork.nodes[paper]['Colour'] = 'orange'
    else:
        citationsNetwork.nodes[paper]['Importance'] = 150
        citationsNetwork.nodes[paper]['Colour'] = 'red'
```

```python
for citation in citationsNetwork.edges():
    if citationsNetwork.nodes[citation[0]]['Importance'] == 600:
        citationsNetwork[citation[0]][citation[1]]['Colour'] = 'green'
        citationsNetwork[citation[0]][citation[1]]['Size'] = 1.5
    elif citationsNetwork.nodes[citation[0]]['Importance'] == 350:
        citationsNetwork[citation[0]][citation[1]]['Colour'] = 'orange'
        citationsNetwork[citation[0]][citation[1]]['Size'] = 1.0
    else:
        citationsNetwork[citation[0]][citation[1]]['Colour'] = 'red'
        citationsNetwork[citation[0]][citation[1]]['Size'] = 0.5
```

```python
plt.figure(figsize = (15, 7))
pos = nx.spring_layout(citationsNetwork)
nx.draw_networkx_nodes(citationsNetwork, pos, node_shape = 'o', #with_labels = True,
                       node_size = [importance for importance in nx.get_node_attributes(citationsNetwork, 'Importance').values()],
                       node_color = [color for color in nx.get_node_attributes(citationsNetwork, 'Colour').values()], 
                       alpha = 0.7)
nx.draw_networkx_edges(citationsNetwork, pos, 
                       width = [size for size in nx.get_edge_attributes(citationsNetwork, 'Size').values()],
                       edge_color = [color for color in nx.get_edge_attributes(citationsNetwork, 'Colour').values()],
                       alpha = 0.5)
nx.draw_networkx_labels(citationsNetwork, pos, font_size = 9)
plt.savefig("citations.png")
```
